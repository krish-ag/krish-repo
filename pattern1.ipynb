### Pyramid Pattern

```python
def pattern(n):
    k=2 * n - 2
    for i in range(0,n):
        for j in range(0,k):
            print(end=' ')
        k=k-1
        for j in range(0,i+1):
            print("*",end=" ")
        print('\r')
        
pattern(int(input("enter a no.")))
```

### Inverse Pyramid Pattern

```python
n=int(input('Enter a no.'))
k=(2*n-2)
for i in range(n,-1,-1):
    for j in range(k,0,-1):
        print(end=' ')
    k=k+1
    for j in range(0,i+1):
        print('*',end=' ')
    print("\r")
```

### Right Start Pattern

```python
def pattern(n):
    for i in range(0,n):
        for j in range(0,i+1):
            print('*',end='')
        print('\r')
    for i in range(n,-1,-1):
        for j in range(0,i+1):
            print('*',end='')
        print('\r')
pattern(10)
```

### Left Start Pattern

```python
def pattern(n):
    k=2*n-2
    for i in range(0,n-1):
        for j in range(0,k):
            print(end=' ')
        k=k-2
        for j in range(0,i+1):
            print('* ',end='')
        print('\r')
    k= -1
    for i in range(n-1,-1,-1):
        for j in range(k,-1,-1):
            print(end=' ')
        k=k+2
        for j in range(0,i+1):
            print('* ',end='')
        print('\r')
pattern(int(input('enter a no.')))
```

### Hour Glass

```python
def pattern(n):
    k=n - 2
    for i in range(n,-1,-1):
        for j in range(k,0,-1):
            print(end=' ')
        k=k+1
        for j in range(0,i+1):
            print('*',end=' ')
        print("\r")
    k=2*n-2
    for i in range(0,n+1):
        for j in range(0,k):
            print(end=' ')
        k=k-1
        for j in range(0,i+1):
            print("*",end=" ")
        print('\r')
pattern(5)
```

### Diamond Pattern 

```python
def pattern(n):
    k=2*n-2
    for i in range(0,n):
        for j in range(0,k):
            print(end=" ")
        k=k-1
        for j in range(0,i+1):
            print('0',end=' ')
        print('\r')
    k=n-2
    for i in range(n,-1,-1):
        for j in range(k,0,-1):
            print(end=' ')
        k=k+1
        for j in range(i,-1,-1):
            print('0',end=' ')
        print('\r')
pattern(5)
```

### Diamond star pattern

```python
def pattern(n):
    for i in range(n):
        for j in range(n):
            if i+j==2 or i-j==2 or j+i==6 or j-i==2:
                print('*',end='')
            else:
                print(end=' ')
        print()
pattern(5)
```

## Number Pattern Programs

### Number Triangle

```python
def pattern(n):
    x=0
    for  i in range(0,n):
        x+=1
        if x<10:
            for j in range(0,i+1):
                print(("0"+str(x)),end=" ")
            print('\r')
        if x>9:
            for j in range(i+1):
                print(x,end=' ')
            print('\r')
            
pattern(10)
```

### Pascal's Triangle

```python
def pattern(n):
    for i in range(0,n):
        for j in range(0,i+1):
            print(function(i,j),' ',end='')
        print()
def function(n,k):
    r=1
    if (k>n-k):
        k=n-k
    for i in range(0,k):
        r*=n-i
        r//=i+1
    return r
pattern(10)
```

### Descending Order

```python
def pattern(n):
    for i in range(n,0,-1):
        for j in range(1,i+1):
            print(j,end=' ')
        print('\r')
pattern(5)
```

### Binary patterns

```python
def pattern(n):
    k=2*n-2
    for i in range(0,n):
        for j in range(0,k):
            print(end=' ')
        k=k-1
        for j in range(0,i+1):
            print(10,end='')
        print('\r')
pattern(5)
```

## Character Pattern Programs

### Right Alphabetical Triangle

```python
def pattern(n):
    x=65
    for i in range(0,n):
        ch=chr(x)
        x=x+1
        for j in range(0,i+1):
            print(ch,end=' ')
        print('\r')
pattern(26)
```
